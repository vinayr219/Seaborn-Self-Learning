# Basics

```python
import pandas as pd 
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print ("Setup Complete")
```

```python
athletes=pd.read_csv('./Dataset/athlete_events.csv')
athletes.head()
```

```python
athletes.Age=athletes.Age.fillna(athletes.Age.mean())
athletes.Height=athletes.Height.fillna(athletes.Height.mean())
athletes.Weight=athletes.Weight.fillna(athletes.Weight.mean())
athletes.Medal=athletes.Medal.fillna("Unknown")
athletes.head()
```

# Line Plot
Line Plot Example :
    First set width and height of the fig and then plot it

```python
plt.figure(figsize=(20,10))

sns.lineplot(data=athletes.loc[0:499,['Age','Height','Weight']])
```

# Bar Charts

```python
athletes.set_index('Name')
```

```python
ath1=athletes.loc[0:299,['Name','Age','Height','Weight','Medal']]
ath1.set_index('Name')
```

```python
plt.figure(figsize=(30,10))
sns.barplot(x=ath1['Height'],y=ath1['Weight'])
```

# HeatMap


```python
ath1.index_col=0
ath1.head()
```

```python
plt.figure(figsize=(3,5))

sns.heatmap(data=ath1.loc[:10,['Age','Weight']],annot=True)
#Heatmap plots index at y-axis and columns at x-axis and values as color coded boxes
```

# ScatterPlots

```python

plt.figure(figsize=(12,5))

sns.scatterplot(x=ath1.Height,y=ath1.Weight)
```

```python
plt.figure(figsize=(12,5))
sns.regplot(x=ath1.Height,y=ath1.Weight)
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x=ath1.Height,y=ath1.Weight,hue=ath1.Medal)
```

```python
plt.figure(figsize=(20,5))

sns.lmplot(x='Height',y='Weight',hue='Medal',data=ath1)
#the colored lines are regression lines
```

```python
#swarmplot , also referred as , categorical scatter plot

plt.figure(figsize=(12,5))

sns.swarmplot(x=ath1.Medal,y=ath1.Height)
```

# Distribution :

### 1.Histogram


```python
plt.figure(figsize=(12,5))

sns.distplot(a=ath1.Age,kde=False)
sns.distplot(a=ath1.Height,kde=False)
```

```python
sns.distplot(a=ath1.Age,kde=True)
#kde= Kernel Density Estimate
```

### 2. Density Plots

```python
plt.figure(figsize=(12,5))

sns.kdeplot(data=ath1.Height,shade=True)
sns.kdeplot(data=ath1.Weight,shade=True)
sns.kdeplot(data=ath1.Age,shade=True)
```

### 3. 2D KDE

#### JointPlot


```python
plt.figure(figsize=(12,5))

sns.jointplot(x=ath1.Age,y=ath1.Height,kind='kde')
```
